```python
# Importamos librerías (por seguridad)
from rdkit import Chem
from rdkit.Chem import Draw

# Lista de SMILES
smiles_list = ["CCO", "c1ccccc1", "CC(=O)O", "c1ccccc1N"]

# Convertimos los SMILES en moléculas
molecules = [Chem.MolFromSmiles(sm) for sm in smiles_list]

# Dibujamos las moléculas en 2D
Draw.MolsToGridImage(molecules, molsPerRow=2, subImgSize=(300,300))
```
